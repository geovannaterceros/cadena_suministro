
## CONTROL ESTADÍSTICO DE PROCESOS EN PYTHON(pyspc)

Introducción

PySpc es una biblioteca de Python destinada a hacer que los Gráficos de control de procesos estadísticos sean lo más fáciles posible.[1]

Caracteristicas de las Gráficos de control por variables:

* Media y amplitud
* Desviación media y estándar
* Valores individuales y rango de movimiento
* Valores individuales con subgrupos
* Promedio móvil ponderado exponencialmente (EWMA)
* Suma acumulativa (CUSUM)

## 1. Utilizando el promedio Movil ponderado exponencial (EWMA)

```python
from pyspc import *
import numpy as np
import pandas as pd
```

El Dataset es relacionado a un fraude en un banco.[2]

```python
fraude_df = pd.read_csv('tc.csv')
fraude_df.head(3)
```

Se utilizar la columna de Cantidad de ingresos(Amount) para aplicar las caracteristicas

```python
a = spc(fraude_df['Amount']) + ewma()
print(a)
```

```python
#Puntos criticos son las puntos rojos
print(a + rules())
```

## 2. Utilizando la suma acumulativa (CUSUM)

```python
print(spc(fraude_df['Amount'])+ cusum())
```
